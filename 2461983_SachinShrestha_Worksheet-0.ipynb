```python
#7
#TASK1

# Daily time (in hours): [study, entertainment, sleep]
time_data = [
 (3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
 (4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
 (5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
 (3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
 (4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
 ]

low=[]
moderate=[]
high=[]

for data in time_data:
    if data[0] < 3.0:
        low.append(data[0])
    elif data[0] >= 3.0 and data[0] <= 5.0:
        moderate.append(data[0])
    else:
        high.append(data[0])


print("Low study hours:", low)
print("Moderate study hours:", moderate)
print("High study hours:", high)
        
```

```python
#8.1
#TASK1

# sample nested list
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]

def sum_nested_list(nested_list):
    total = 0
    for element in nested_list:
        if type(element) == list:
            total += sum_nested_list(element)
        else:
            total += element
    return total

total = sum_nested_list(nested_list)
print("Sum of all numbers:", total)
```

```python
#TASK2

def generate_permutations(s):
    if len(s) == 0:
        return [""]
    permutations = set()
    for i in range(len(s)):
        current = s[i]
        remaining = s[:i]+s[i+1:]
        for p in generate_permutations(remaining):
            permutations.add(current + p)
    return permutations

print(list(generate_permutations("abc")))
print(list(generate_permutations("aab")))
```

```python
#8.2
#TASK1

def min_coins(coins, amount):
    dp = [float("inf")] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    if dp[amount] != float("inf"):
        return dp[amount]
    else:
        return -1

print(min_coins([5,5,1], 11))
```

```python
#TASK2

def longest_common_subsequence(s1, s2):
    m = len(s1)
    n = len(s2)
    dp = []
    for i in range(m + 1):
        dp.append([0] * (n + 1))

    for j in range(1, m + 1):
        for k in range(1, n + 1):
            if s1[j-1] == s2[k-1]:
                dp[j][k] = dp[j-1][k-1] + 1

            else:
                dp[j][k] = max(dp[j-1][k], dp[j][k-1])

    return dp[m][n]
print(longest_common_subsequence("abcde", "ace"))
```

```python
#TASK3

def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]
    
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(values[i - 1] + dp[i - 1][w - weights[i - 1]], dp[i - 1][w])
            else:
                dp[i][w] = dp[i - 1][w]
    
    return dp[n][capacity]

weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
print(knapsack(weights, values, capacity))
 
```
