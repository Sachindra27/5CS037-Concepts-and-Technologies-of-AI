```python
import pandas as pd
import numpy as np
df = pd.read_csv('Titanic-Dataset.csv')
df.info()
```

```python
#Subsetting

#1
fare_gt_100 = df[df['Fare'] > 100]
print(f"Passengers with fare > 100: {len(fare_gt_100)}")
```

```python
#2
first_class = df[df['Pclass'] == 1]
print(f"Passengers in First Class: {len(first_class)}")
```

```python
#3
female_under_18 = df[(df['Age'] < 18) & (df['Sex'] == 'female')]
print(f"Females under 18: {len(female_under_18)}")
```

```python
# Subsetting Rows by Categorical Variables

#1
embarked_c_or_s = df[df['Embarked'].isin(['C', 'S'])]
print(f"Embarked from C or S: {len(embarked_c_or_s)}")
```

```python
#2
first_second_class = df[df['Pclass'].isin([1, 2])]
print(f"Passengers in 1st or 2nd Class: {len(first_second_class)}")
```

```python
#EDA 3.2
median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)
print(f"Missing age values filled with median value: {median_age}")
```

```python
#Q1

#1
df['fare_per_year'] = df['Fare'] / df['Age']

#2
high_fare_age = df[df['fare_per_year'] > 5]

#3
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)

#4
result_q1 = high_fare_age_srt[['Name', 'fare_per_year']]

#5
print("Top fare per year:")
print(result_q1.head())
```

```python
#Q2

#1
df['fare_per_class'] = df['Fare'] / df['Pclass']

#2
adult_males = df[(df['Sex'] == 'male') & (df['Age'] >= 18)]

#3
adult_males_srt = adult_males.sort_values(by='fare_per_class', ascending=False)

#4
result_q2 = adult_males_srt[['Name', 'Age', 'fare_per_class']]

#5
print("Adult males who paid the highest fares:")
print(result_q2.head())
```

```python
#EDA 3.3

#Q1

#1
total_fare = df['Fare'].sum()

#2
fare_c1 = df[df['Pclass'] == 1]['Fare'].sum()

#3
fare_c2 = df[df['Pclass'] == 2]['Fare'].sum()
fare_c3 = df[df['Pclass'] == 3]['Fare'].sum()

#4
fare_totals = [fare_c1, fare_c2, fare_c3]

#5
fare_proportions = [fare / total_fare for fare in fare_totals]

classes = ['First class', 'Second class', 'Third class']
for cls, prop in zip(classes, fare_proportions):
    print(f"{cls} fare percentage: {prop:.2%}")
```

```python
#Q2

#1
def get_age_group(age):
    if age < 18:
        return 'child'
    elif 18 <= age <= 64:
        return 'adult'
    else:
        return 'senior'

df['age_group'] = df['Age'].apply(get_age_group)

#2
total_passengers = len(df)

#3
group_counts = df['age_group'].value_counts()

#4
passengers_proportions = (group_counts / total_passengers)

#5
proportion_percentage = passengers_proportions * 100
print("Percentages of age groups from total passengers:")
print(proportion_percentage)
```
