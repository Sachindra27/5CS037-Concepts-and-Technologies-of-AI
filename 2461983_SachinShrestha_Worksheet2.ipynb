```python
#Problem 2
#1

medical_file_path = "medical_students_dataset.csv"
med_df = pd.read_csv(medical_file_path)
med_df
```

```python
#2

print(med_df.info())
null_columns = med_df.columns[med_df.isnull().any()].tolist()
print("\nColumns with missing values:")
print(null_columns)
```

```python
#3

mean_fill_cols = ['Age', 'Height', 'Weight', 'BMI', 'Temperature', 'Heart Rate']
for col in mean_fill_cols:
    med_df[col] = med_df[col].fillna(med_df[col].mean())

median_fill_cols = ['Cholesterol', 'Blood Pressure']
for col in median_fill_cols:
    med_df[col] = med_df[col].fillna(med_df[col].median())

mode_fill_cols = ['Gender', 'Blood Type', 'Diabetes', 'Smoking']
for col in mode_fill_cols:
    med_df[col] = med_df[col].fillna(med_df[col].mode()[0])
```

Explanation:

I. Mean fill: The numerical columns Age, Height, Weight, BMI, Temperature and Heart Rate have continuous values and symmetric distributions hence mean fill is used.

II. Median fill: The numerical columns Cholesterol and Blood Pressure have skewed distributions hence median fill is used.

III. Mode fill: The columns Gender, Blood Type, Diabetes and Smoking are categorical in nature hence mode fill is used to fill in the null spaces with the most common category.

```python
#4

num_duplicate = med_df.duplicated().sum()
print("Duplicate rows count:", num_duplicate)
if num_duplicate > 0:
    duplicate_removed = med_df.drop_duplicates()
    print("Dropped duplicates.")
else:
    print("No duplicates found.")
```
